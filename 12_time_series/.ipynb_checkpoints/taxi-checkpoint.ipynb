## Обучение

```python
def make_features(df, max_lag, rolling_mean_size):
    df['month'] = df.index.month
    df['day'] = df.index.day
    df['dayofweek'] = df.index.dayofweek
    df['hour'] = df.index.hour
    
    for lag in range(1, max_lag + 1):
        df['lag_{}'.format(lag)] = df['num_orders'].shift(lag)

    df['rolling_mean'] = df['num_orders'].shift().rolling(rolling_mean_size).mean()
    
make_features(df, 24, 100)
print(df.head())
```

Подбор гиперпараметров условный - чтобы удовлетворял требованиям и занимал минимальное время

```python
train, test = train_test_split(df, shuffle=False, test_size=0.1)
train = train.dropna()
test = test.dropna()

X_train = train.drop(['num_orders'], axis=1)
y_train = train['num_orders']
 
X_test = test.drop(['num_orders'], axis=1)
y_test = test['num_orders']

my_cv = TimeSeriesSplit(n_splits=5)
```

#### LinearRegression

```python
model = LinearRegression()

param_grid = {
    'fit_intercept': [False],
    'normalize': [True]
}

grid_search = GridSearchCV(model, param_grid=param_grid, cv=my_cv, scoring='neg_root_mean_squared_error')
grid_search.fit(X_train, y_train)

best_params_lr = grid_search.best_params_
print(f'best_params: {best_params_lr}')

results = grid_search.cv_results_
best_index = grid_search.best_index_

fit_time = results['mean_fit_time']
score_time = results['mean_score_time']
RMSE = grid_search.best_score_ * -1

print(f'Время обучения: {fit_time}, Время предсказания: {score_time}, RMSE: {RMSE}')
```

#### LGBMRegressor

```python
model = lgbm.LGBMRegressor()
param_grid = {
    'learning_rate': [0.3],
    'n_estimators': [150]
}

grid_search = GridSearchCV(
    model,
    param_grid=param_grid,
    cv=my_cv,
    scoring='neg_root_mean_squared_error',
    n_jobs=-1,
)

grid_search.fit(X_train, y_train)

best_params = grid_search.best_params_
best_params_lgbm = grid_search.best_params_
print(f'best_params: {best_params_lgbm}')

results = grid_search.cv_results_
best_index = grid_search.best_index_

fit_time = results['mean_fit_time']
score_time = results['mean_score_time']
RMSE = grid_search.best_score_ * -1

print(f'Время обучения: {fit_time}, Время предсказания: {score_time}, RMSE: {RMSE}')
```

#### CatBoostRegressor

```python

model = CatBoostRegressor()
param_grid = {
    'iterations': [30],
    'learning_rate': [0.3],
}

grid_search = GridSearchCV(
    model,
    param_grid,
    cv=my_cv, scoring='neg_root_mean_squared_error',
    error_score='raise',
    n_jobs=-1,
)
grid_search.fit(X_train, y_train)

best_params_cbr = grid_search.best_params_
print(f'best_params: {best_params_cbr}')

results = grid_search.cv_results_
best_index = grid_search.best_index_

fit_time = results['mean_fit_time']
score_time = results['mean_score_time']
RMSE = grid_search.best_score_ * -1

print(f'Время обучения: {fit_time}, Время предсказания: {score_time}, RMSE: {RMSE}')
```

Лучший RMSE показал CatBoostRegressor - 25.39 - проверим эту модель на тестовой выборке:

## Тестирование

Проверим лучшую модель на тестовой выборке:

```python
model = CatBoostRegressor(**grid_search.best_params_)
model.fit(X_train, y_train)

# Предсказание на тестовой выборке
y_pred = model.predict(X_test)

# Оценка качества модели
print('RMSE тестовой выборки лучшей модели: ', mean_squared_error(y_test, y_pred)**0.5)
```

Проверим на адекватность:

```python
pred_previous = y_test.shift()
pred_previous.iloc[0] = y_train.iloc[-1]
print('RMSE:', mean_squared_error(y_test, pred_previous)**0.5)
```

Все модели, на которых проводилось обучение, выдали необходимый результат по метрике RMSE. Лучший результат показала модель CatBoostRegressor с результатом RMSE - 44.
